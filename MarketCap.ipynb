```python
import pandas as pd
import yfinance as yf

# Example: list of tickers (replace with your Excel reading code)
file_path = 'SP 500 List.csv'
tickers = pd.read_csv(file_path, header=None).iloc[0].dropna().tolist()


# Create empty DataFrame to store market caps
all_market_caps = pd.DataFrame()

for ticker in tickers:
    try:
        # Download monthly close data
        data = yf.download(ticker, start='2000-01-01', interval='1mo', progress=False)
        
        # Get shares outstanding (current)
        info = yf.Ticker(ticker).info
        shares_outstanding = info.get('sharesOutstanding', None)

        if shares_outstanding is None:
            print(f"⚠️  Shares outstanding not found for {ticker}")
            continue
        
        # Calculate Market Cap = Close * Shares Outstanding
        data['Market Cap'] = data['Close'] * shares_outstanding
        
        # Keep only Market Cap and rename column
        data = data[['Market Cap']].rename(columns={'Market Cap': ticker})
        
        # Merge into main DataFrame
        all_market_caps = pd.concat([all_market_caps, data[ticker]], axis=1)
    
    except Exception as e:
        print(f"Error downloading {ticker}: {e}")

# Sort by date
all_market_caps = all_market_caps.sort_index()
all_market_caps.index.name = 'Date'

# Save to Excel
all_market_caps.to_csv('monthly_market_caps.csv')

print("✅ Done! Monthly market caps saved to 'monthly_market_caps.xlsx'")
```
