```python
import pandas as pd
import yfinance as yf

# Example: list of tickers (replace with your Excel reading code)
file_path = 'SP 500 List.csv'
tickers = pd.read_csv(file_path, header=None).iloc[0].dropna().tolist()

# Create empty DataFrame to store monthly average volumes
all_avg_volumes = pd.DataFrame()

for ticker in tickers:
    try:
        # Download monthly data
        data = yf.download(ticker, start='2000-01-01', interval='1mo', progress=False)
        
        # Only keep the "Volume" column
        data = data[['Volume']].rename(columns={'Volume': ticker})
        
        # Combine into main DataFrame
        all_avg_volumes = pd.concat([all_avg_volumes, data[ticker]], axis=1)
        
    except Exception as e:
        print(f"Error downloading {ticker}: {e}")

# Sort by date
all_avg_volumes = all_avg_volumes.sort_index()
all_avg_volumes.index.name = 'Date'

# Save to Excel
all_avg_volumes.to_csv('monthly_average_volumes.csv')

```
